```python
# import library 

import pandas as pd
import numpy as np
```

```python
# load dataset

cc_fraud = pd.read_csv('../data/credit_card_fraud_dataset.csv')
```

```python

# menampilkan 5 baris pertama data 

cc_fraud.head()
```

```python
# ubah TransactionDate menjadi datetime
cc_fraud['TransactionDate'] = pd.to_datetime(cc_fraud['TransactionDate'], errors='coerce')

cc_fraud.head()
```

```python
# cek info apakah sudah terubah dtype nya menjadi datetime.
cc_fraud.info()
```

```python
# membuat sebuah kolom HourofDay dan kolom DayofWeek.

cc_fraud['HourofDay'] = cc_fraud['TransactionDate'].dt.hour
cc_fraud['DayofWeek'] = cc_fraud['TransactionDate'].dt.day_of_week
cc_fraud.head()
```

```python
# membuat sebuah kolom boolean Is_Night_Transaction yang bernilai True jika transaksi terjadi antara jam 00:00 hingga 05:00 pagi (waktu rawan fraud).

cc_fraud['Is_Night_Transaction'] = cc_fraud['HourofDay'] < 6
cc_fraud.head(10)
```

```python
# membuat sebuah  kolom Risk_Level dengan kategori Critical, Medium, Low
cc_fraud['Risk_Level'] = np.where((cc_fraud['Amount'] > 4000) & (cc_fraud['TransactionType'] == "refund"), 'Critical', np.where(cc_fraud['Amount'].between(2000,4000),'Medium','Low'))
```

```python
cc_fraud.head(10) # menampilkan 10 baris awal data
```

```python
# Buat kolom Flag_Suspicious.
# Berikan nilai 1 jika transaksi terjadi di 'New York' atau 'Houston' dengan Amount di atas rata-rata (mean) seluruh transaksi, jika tidak 0.

mean_amount = cc_fraud['Amount'].mean()

# Cara (Vectorized)
cc_fraud['Flag_Suspicious'] = np.where(
    (cc_fraud['Location'].isin(["New York", "Houston"])) & (cc_fraud['Amount'] > mean_amount),
    1, 
    0
)

# menampilkan dataset
cc_fraud
```

```python
cc_fraud.query('Flag_Suspicious == 1') # menampilkan data dengan suspecious 1 
```

```python
# membuat sebuah kolom locatio code
cc_fraud['Location_Code'] = cc_fraud['Location'].str[0:3].str.upper()


# menampilkan dataset
cc_fraud
```

```python
# menghapus kolom MerchantID

cc_fraud.drop(columns=['MerchantID'], inplace=True)

cc_fraud
```

```python
# rename kolom IsFraud menjadi Target_Label
cc_fraud.rename(columns={'IsFraud':'Target_Label'}, inplace=True)

cc_fraud
```

```python
# save update dataframe
cc_fraud.to_csv('../data/fraud_analysis_final_2026.csv')
```
