```python
# challenge 1

import numpy as np

rng = np.random.default_rng(seed=42)
market_data = rng.integers(low=100, high=200, size=(10, 3))
print(market_data)
```

```python
# challenge 2 (analisis statistik)

# Hitung rata-rata (mean) untuk setiap kolom open, high, low, close.
print(np.mean(market_data, axis = 0))

# cari harga tertinggi secara global
print(np.max(market_data))

# hitung standart deviasi (std) untuk melihat volatilitas pasar secara kesuluruhan
print(np.std(market_data))
```

```python
# challenge 3

# penyesuain harga (broadcasting)
# kurangi seluruh isi market data dengan angka 2.5
# pembulatan hasil 2 angka dibelakang koma
print(np.round(market_data - 2.5, 2))
```

```python
# challenge 4
# filtering & multiple conditions
hari_bagus = market_data[(market_data[:,0]>180) & (market_data[:,2]<150)]
print(hari_bagus)
```

```python
# challenge 5
# buat array bernama signal 
signal = np.array(np.where(market_data[:,2] > 170, "SELL", "HOLD"))
print(signal)

```

```python
# challenge 6
# Sorting Close dari market_data terbesar ke terkecil
sorting_descending = np.sort(market_data[:, 2])[::-1]
print(sorting_descending)

```
