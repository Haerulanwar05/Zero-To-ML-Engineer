# Praktik Slicing di Numpy

```python
# 1D (Vector) Slicing

# membuat array 1d dengan data Harga Penutupan Nasdaq
price_close = np.array([14944.35,14972.76,14970.18,14969.65,14857.71,14843.77,14524.07,14510.30,14592.21,14765.94])

# Slicing harga dari indeks ke-3 sampai ke-7
print("Harga indeks - 3 sampe indeks - 7")
print(price_close[3:8])

# Pembalikan Data menggunakan step
print("Pembalikan Data")
print(price_close[::-1])
```

```python
# 2D (Matriks) Slicing

# membuat array 2D (5x4) dengan data harga nasdaq seminggu
price_week_nasdaq = np.array([
    [14129.96,14149.62,13982.42,14031.81],
    [13994.54,14060.85,13945.65,14020.95],
    [13988.81,14001.95,13802.42,13872.47],
    [13675.11,13774.63,13642.47,13748.83],
    [13754.71,13843.37,13733.18,13761.53]])

# Ekstrak hanya kolom High (indeks 1) dan Low (indeks 2) untuk semua baris.
print("High (indeks 1) dan Low (indeks 2) untuk semua baris.")
print(price_week_nasdaq[: ,1:3])

# Ambil hanya 3 baris terakhir untuk semua kolom.
print("3 baris terakhir untuk semua kolom.")
print(price_week_nasdaq[2:])

# negatif indexing Ambil data Close (kolom terakhir) pada Baris terakhir
print("data Close (kolom terakhir) pada Baris terakhir")
print(price_week_nasdaq[-1: ,-1])
```

# Praktik Filtering di Numpy

```python
import numpy as np

# chalengge 1 

# data nilai uas
nilai_uas = np.array([80, 65, 90, 50, 76, 88, 40])

# filter nilai yang lulus diatas 75 dari data nilai_uas
lulus = nilai_uas[nilai_uas > 75]

# menampilkan nilai yang lulus diatas 75
print("Nilai yang lulus: ")
print(lulus)

# filter nilai yang remidi dibawah 75 dari data nilai_uas
remidi = nilai_uas[nilai_uas < 75]

# menampilkan nilai yang remidi dibawah 75
print("Nilai yang remidi: ")

print(remidi)


```

```python
# chalengge 2 

# membuat sebuah variabel berisi data hasil trading 
data_trading = np.array([-100, 200, 50, -50, 400, 0, -20])

# filtering hasil trading profit > 0
profit_plus = data_trading[data_trading > 0]
# menampilkan hasil filtering 
print("hasil trading profit (+): ")
print(profit_plus)

# filtering hasil trading profit lebih dari 100
huge_profit = data_trading[data_trading > 100]
# menampilkan hasil filtering
print("hasil trading profit lebih dari 100: ")
print(huge_profit)

# filtering hasil trading loss atau Break Even
loss_BE = data_trading[(data_trading < 0) | (data_trading == 0)]
# menampilkan hasil Filtering
print("hasil trading Loss/Be: ")
print(loss_BE)

# filtering hasil trading profit dibawah 100
profit_under100 = data_trading[(data_trading > 0) & (data_trading < 100)]
# menampilkan hasil Filtering
print("hasil trading Profit < 100: ")
print(profit_under100)
```

```python
# challenge 3

# penggunaan np.where ()

# data hasil trading 
hasil_trading = np.array([10, -5, 20, 0, -15, 30])

# filter data dan melabeli data
print(np.where(hasil_trading > 0, "Buy", "Wait"))
```
