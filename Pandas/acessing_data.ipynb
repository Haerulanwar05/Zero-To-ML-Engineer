# Akses Data dengan Library Pandas

```python
import pandas as pd
```

```python
# load dataset dengan variabel yang dibuat
coffe = pd.read_csv('data/coffee.csv')
```

```python
coffe.head()        # untuk melihat 5 baris pertama dari data coffee 
```

```python
coffe       # untuk melihat semua dataframe 
```

```python
# Membuat sebuah data random atau sebuah sample dari data coffe
coffe.sample(10, random_state=1)
```

```python
# Akses Data dengan loc dan iloc
coffe 
```

```python
# memberi tahu value dari dataframe berdasarkan baris yang ingin di ketahui dengan sebuah label atau nama index atau column bisa dengan values nya atau label nya


coffe.loc[1] 
```

```python
coffe.loc[[1,2,6]]
```

```python
# dengan sebuah loc kita bisa tau ingin mengambil value apa dari sebuah dataframe.
coffe.loc[:,["Day","Units Sold","Coffee Type"]]     
```

```python
# sama dengan loc bedanya iloc harus menggunakan angka.
coffe.iloc[:,[0,2]]     
```

```python
coffe
```

```python
# mengatur sebuah dataframe values dengan loc/iloc

coffe.loc[1:3, "Units Sold"] = 10
```

```python
coffe
```

### mengambil sebuah kolom, memfilter (sort) nilai, ascending - descending

```python
coffee_new = pd.read_csv('data/coffee.csv')
```

```python
coffee_new
```

```python
# sort values dari dataframe coffee
coffee_new.sort_values(by='Units Sold')
```

```python
# sort values dari dataframe coffee dari terbesar ke terkecil
coffee_new.sort_values(['Units Sold','Coffee Type'], ascending=False)
```

```python
# untuk mengambil sebuah kolom
coffee_new['Coffee Type']
```

# Praktek

```python
praktik_coffee = pd.read_csv('data/coffee.csv')
```

```python
praktik_coffee
```

```python
# 4 baris pertama Dataframe 
praktik_coffee.head(4)
```

```python
# 2 baris terakhir DataFrame
praktik_coffee.tail(2)
```

```python
# 3 baris acak dari data DataFrame 
praktik_coffee.sample(3)
```

```python
# menggunakan iloc mengambil data baris ke - 5 sampai ke - 8
praktik_coffee.iloc[5:9]
```

```python
# menggunakan iloc untuk mengambil jumlah 'units sold' pada 5 baris pertama
praktik_coffee.iloc[0:6,2]
```

```python
# Gunakan loc untuk menampilkan baris indeks 10 sampai 12, tapi hanya kolom 'Day' dan 'Units Sold'.
praktik_coffee.loc[10:12,["Day","Units Sold"]]
```

```python
# jumlah unit yang terjual di baris indeks 0? Ambil nilainya saja pakai iloc.
praktik_coffee.iloc[0,2]
```
