# Merging & Concetenating Data

```python
# Import Library

import pandas as pd
```

```python
# Load Dataset bios

bios = pd.read_csv('data/bios.csv')
```

```python
bios.head()
```

```python
# Load Dataset ke-2 
nocs = pd.read_csv('data/noc_regions.csv')
nocs.head()
```

```python
# merge Dataframe 
bios_new = pd.merge(bios, nocs, left_on='born_country', right_on='NOC', how='left') 
```

```python
bios_new.head()
```

```python
# rename kolom 'region' dengan 'born_country_full

bios_new.rename(columns={'region':'born_country_full'}, inplace=True)
```

```python
bios_new.head()
```

di dalam sebuah dataframe baru 'bios_new' hasil dari merging terdapat kolom yang bernama NOC_x dan NOC_y, ini terjadi karena di dalam kedua Dataframe ini terdapat Kolom yang Sama namun dengan isi yang berbeda.

```python
# contoh concatenate

# hanya mengambil bio athlete dari USA saja
USA = bios[bios['born_country'] == "USA"].copy()

# hanya mengambil bio athlete dari gbr saja
GBR = bios[bios['born_country'] == "GBR"].copy()
```
