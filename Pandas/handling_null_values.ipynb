# Handling null values

```python
# import library 

import pandas as pd
import numpy as np
```

```python
# load dataframe

coffe = pd.read_csv('data/coffee.csv')
```

```python
coffe.head()
```

```python
# buat kolom price dimana Espresso = 3.99 dan Latte = 5.99

coffe['price'] = np.where(coffe['Coffee Type'] == "Espresso", 3.99, 5.99)
```

```python
coffe.head()
```

```python
# buat kolom revenue
coffe['revenue'] = coffe['Units Sold'] * coffe['price']
```

```python
coffe.head()
```

```python
# membuat data indexs baris 0 dan 1 menjadi NaN

coffe.loc[[0,1], 'Units Sold'] = np.nan
```

```python
coffe.head()
```

```python
# cara cek data memiliki suati nilai atau bernilai NaN
coffe.info()


# di Units Sold terdapat 12 non-null dimana berarti ada 2 null di kolom tersebut
```

```python
# opsi lain mencari data NaN
coffe.isna().sum()

# terdapat 2 data null di kolom Units Sold
```

```python
# mengisi data null dengan mean kolom 'Units Sold' dengan fungsi .fillna()

# coffe.fillna(coffe['Units Sold'].mean())
```

```python
coffe.head()
```

```python
# mengosongkan data dengan NaN di indexs baris 2 dan 3
coffe.loc[[2,3], 'Units Sold'] = np.nan
```

```python
coffe
```

```python
coffe.loc[[0,1],'Units Sold'] = 15
```
