# Handling null values

```python
# import library 

import pandas as pd
import numpy as np
```

```python
# load dataframe

coffe = pd.read_csv('data/coffee.csv')
```

```python
coffe.head()
```

```python
# buat kolom price dimana Espresso = 3.99 dan Latte = 5.99

coffe['price'] = np.where(coffe['Coffee Type'] == "Espresso", 3.99, 5.99)
```

```python
coffe.head()
```

```python
# buat kolom revenue
coffe['revenue'] = coffe['Units Sold'] * coffe['price']
```

```python
coffe.head()
```

```python
# membuat data indexs baris 0 dan 1 menjadi NaN

coffe.loc[[0,1], 'Units Sold'] = np.nan
```

```python
coffe.head()
```

```python
# cara cek data memiliki suati nilai atau bernilai NaN
coffe.info()


# di Units Sold terdapat 12 non-null dimana berarti ada 2 null di kolom tersebut
```

```python
# opsi lain mencari data NaN
coffe.isna().sum()

# terdapat 2 data null di kolom Units Sold
```

```python
# mengisi data null dengan mean kolom 'Units Sold' dengan fungsi .fillna()

# coffe.fillna(coffe['Units Sold'].mean())
```

```python
coffe.head()
```

```python
# mengosongkan data dengan NaN di indexs baris 2 dan 3
coffe.loc[[2,3], 'Units Sold'] = np.nan
```

```python
coffe
```

```python
coffe.loc[[0,1],'Units Sold'] = 15
```

berusaha menjaga tren atau pola yang ada dalam kolom tersebut.

```python
# cara lain menampilkan mana data yang NaN dan tidak NaN


# menampilkan data yang NaN
coffe[coffe['Units Sold'].isna()]
```

# Praktek 

```python
# load dataframe
df_coffee = pd.read_csv('data/coffee_sales.csv')
```

```python
df_coffee.head()
```

```python
df_coffee.info()
```

```python
# menampilkan ringkasan jumlah NaN di setiap Kolom

df_coffee.isna().sum()
```

```python
# menampilkan data dengan nilai kolom card NaN
df_coffee[df_coffee['card'].isna()]
```

```python
# mengosongkan nilai di index 10 - 20 di kolom money
df_coffee.loc[10:20, 'money'] = np.nan

# mengosongkan nilai di index 50-55 di kolom coffee_name
df_coffee.loc[50:55, 'coffee_name'] = np.nan
```

```python
df_coffee.isna().sum()
```

```python
# mengganti nilai NaN di kolom card dengan "CASH_PAYMENT"
df_coffee['card'] = df_coffee['card'].fillna("CASH_PAYMENT")
```

```python
df_coffee[df_coffee['card'] == "CASH_PAYMENT"]
```

```python
# untuk kolom coffee_name, Jika nama kopinya hilang, data itu harus dibuang karena kita tidak tahu stok apa yang berkurang.
df_coffee.dropna(subset='coffee_name',inplace=True)
```

```python
df_coffee.isna().sum()
```

```python
df_coffee['money'] = df_coffee['money'].interpolate()
```

```python
df_coffee
```

```python
# mengecek ulang nilai NaN final
df_coffee.isna().sum()
```

```python
# save data coffee_sales bersih tanpa NaN
df_coffee.to_csv('data/coffee_sales_final_report.csv')
```
