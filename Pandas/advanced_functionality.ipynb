### Fungsi .shift()
Secara sederhana, fungsi ini digunakan untuk menggeser data ke atas atau ke bawah pada indeks tertentu.

```python
# test fungsi .shift()

# membuat kolom baru dengan isi pendapatan kemarin 

coffee['yesterday_revenue'] = coffee['revenue'].shift(2)
```

```python
coffee.head()
```

```python
# membuat pertumbuhan persen

coffee['pct_change'] = coffee['revenue'] / coffee['yesterday_revenue'] * 100
```

```python
coffee
```

### Fungsi .rank()
Fungsi ini digunakan untuk memberikan peringkat atau posisi pada setiap elemen dalam kumpulan data.

```python
# contoh fungsi .rank()

# menerapkan rank dari tinggi suatu atlet 
bios['height_rank'] = bios['height_cm'].rank(ascending=False)
```

```python
# menerapkan sort values dari terbesar ke terkecil
bios.sort_values(['height_rank'])
```

```python
# menampilkan 10 sample dengan mengambil kolom nama dan rank
bios.sort_values(['height_rank']).sample(10)[['name','height_rank']]
```

### Fungsi .rolling()
.rolling() adalah tentang melihat gambaran besar melalui "jendela" yang bergerak.

```python
# menggunakan fungi .rolling()

coffee['cumulative_revenue'] = coffee['revenue'].cumsum()
```

```python
coffee
```

```python
# meng copy data hanya berisi tipe kopi latte dan menghtung kumulatif pendapatan nya per 3 hari menggunakan rolling

latte = coffee[coffee['Coffee Type'] == "Latte"].copy()
latte
```

```python
latte['3day'] = latte['Units Sold'].rolling(3).sum()
```

### apa itu pyarrow?

PyArrow adalah library Python yang menjadi jembatan untuk Apache Arrow, sebuah platform pengembangan lintas bahasa untuk data dalam memori (in-memory data).

PyArrow adalah sebuah pustaka (library) Python yang berfungsi sebagai "mesin penggerak" atau jembatan canggih untuk mengolah data dalam jumlah sangat besar dengan kecepatan tinggi. Bayangkan PyArrow sebagai sistem pengorganisasian data yang sangat rapi di dalam memori komputer, di mana data tidak disimpan baris demi baris seperti daftar belanjaan biasa, melainkan dikelompokkan berdasarkan kolomnya. Cara penyimpanan secara kolom ini membuat komputer jauh lebih cerdas dan efisien saat melakukan perhitungan atau pencarian, karena ia hanya perlu membaca kolom yang benar-benar dibutuhkan tanpa membuang-buang tenaga untuk memproses data yang tidak relevan.

```python
import pandas as pd
import numpy as np

# Membuat data dummy trading
data = {
    'Tanggal': pd.to_datetime(['2024-05-01', '2024-05-02', '2024-05-03', '2024-05-04', '2024-05-05'] * 3),
    'Aset': ['BTC']*5 + ['ETH']*5 + ['SOL']*5,
    'Harga_Close': [60000, 62000, 61000, 63000, 64000,  # BTC
                    3000, 3100, 2900, 3200, 3300,      # ETH
                    140, 150, 145, 160, 155]           # SOL
}

df = pd.DataFrame(data)
# Mengurutkan data berdasarkan Aset dan Tanggal agar logikanya benar
df = df.sort_values(by=['Aset', 'Tanggal']).reset_index(drop=True)
print("Dataset Siap!")
print(df.head(10))
```

```python
df
```

```python
df['Harga_Kemarin'] = df.groupby(['Aset'])['Harga_Close'].shift(1)
```

```python
df
```

```python
df['Perubahan_harga'] = df['Harga_Close'] - df['Harga_Kemarin']
```

```python
df
```

```python
df['rank'] = df['Harga_Close'].rank(method='dense',ascending=False)
```

```python
df.sort_values(['rank'])
```

```python
df['SMA_3'] = df.groupby(['Aset'])['Harga_Close'].rolling(3,min_periods=1).mean().reset_index(0, drop=True)
```

```python
df
```

```python
df['standart_deviasi'] = df.groupby(['Aset'])['Harga_Close'].rolling(3).std().reset_index(0, drop=True)
```

```python
df
```

```python
df['Harga_Besok'] = df.groupby(['Aset'])['Harga_Close'].shift(-1)
```

```python
df
```

```python
df ['signal'] = df['Harga_Close']<df['Harga_Besok']
```

```python
df
```

```python
df['percentil_harga'] = df.groupby(['Aset'])['Harga_Close'].rank(pct=True)
```

```python
df
```

```python
# --- CHALLENGE 9: CROSS-SECTIONAL RANKING (MOMENTUM) ---

# 1. Mencari tanggal terakhir secara otomatis
tanggal_terakhir = df['Tanggal'].max()

# 2. Memfilter data agar hanya berisi baris di tanggal terakhir
# Kita gunakan .copy() agar tidak muncul peringatan SettingWithCopyWarning
df_akhir = df[df['Tanggal'] == tanggal_terakhir].copy()

# 3. Melakukan ranking antar aset berdasarkan kolom SMA_3
# ascending=False -> Nilai SMA tertinggi (tren paling kuat) jadi Nomor 1
df_akhir['Rank_Trend'] = df_akhir['SMA_3'].rank(ascending=False)

# Menampilkan hasil pemenang momentum di hari terakhir
print(f"Hasil Momentum pada Tanggal: {tanggal_terakhir.date()}")
print(df_akhir[['Aset', 'SMA_3', 'Rank_Trend']].sort_values(by='Rank_Trend'))
```
