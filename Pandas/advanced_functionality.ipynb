# Advanced Functionality

```python
# Import Library
import pandas as pd
import numpy as np
```

```python
# Load Dataframe Coffee

coffee = pd.read_csv('data/coffee_update.csv')
bios = pd.read_csv('data/bios.csv')
```

```python
coffee.head()
```

### Fungsi .shift()
Secara sederhana, fungsi ini digunakan untuk menggeser data ke atas atau ke bawah pada indeks tertentu.

```python
# test fungsi .shift()

# membuat kolom baru dengan isi pendapatan kemarin 

coffee['yesterday_revenue'] = coffee['revenue'].shift(2)
```

```python
coffee.head()
```

```python
# membuat pertumbuhan persen

coffee['pct_change'] = coffee['revenue'] / coffee['yesterday_revenue'] * 100
```

```python
coffee
```

### Fungsi .rank()
Fungsi ini digunakan untuk memberikan peringkat atau posisi pada setiap elemen dalam kumpulan data.

```python
# contoh fungsi .rank()

# menerapkan rank dari tinggi suatu atlet 
bios['height_rank'] = bios['height_cm'].rank(ascending=False)
```

```python
menerapkan sort values dari terbesar ke terkecil
bios.sort_values(['height_rank'])
```

```python
# menampilkan 10 sample dengan mengambil kolom nama dan rank
bios.sort_values(['height_rank']).sample(10)[['name','height_rank']]
```

### Fungsi .rolling()
.rolling() adalah tentang melihat gambaran besar melalui "jendela" yang bergerak.

```python
# menggunakan fungi .rolling()

coffee['cumulative_revenue'] = coffee['revenue'].cumsum()
```

```python
coffee
```

```python
# meng copy data hanya berisi tipe kopi latte dan menghtung kumulatif pendapatan nya per 3 hari menggunakan rolling

latte = coffee[coffee['Coffee Type'] == "Latte"].copy()
latte
```

```python
latte['3day'] = latte['Units Sold'].rolling(3).sum()
```
