# Filtering Data | Syntax Operations, Numeric Values, Multiple Conditions

```python
import pandas as pd
```

```python
bios = pd.read_csv('data/bios.csv')
```

```python
bios.head()
```

```python
bios.info()
```

```python
# memfilter tinggi dan mengambil kolom yang diinginkan saja
bios.loc[bios['height_cm'] > 220, ['height_cm', 'name', 'born_country']]    
```

```python
# cara lain
bios[bios['height_cm'] > 200][['height_cm','name']]
```

```python
# jika terdapat 2 kondisi
bios[(bios['height_cm'] > 200) & (bios['born_country']=='USA')].sort_values(['height_cm'])
```

```python
# mencari sebuah nama yang mengandung suatu kalimat yang kita inginkan misal "keith"

# mencari nama keith dengan case = false agar python tidak sensitif dengan angka kecil atau besar
bios[bios['name'].str.contains('keith', case=False)]    
```

```python
# menggunakan sebuah pola pncarian regrex khusus fungsi .str
bios[bios['name'].str.contains('keith|patrick', case=False)]    
```

```python
# jika kita menonaktifkan pola pencarian regrex dalam fungsi .str
# bios[bios['name'].str.contains('keith|patrick', case=False, regex=False)]
```

```python
# cara agar tidak mengulangi kata yang sama seperti bios lalu bios lagi terlihat aneh menggunakan sebuah filter modern agar kode telihat bersih dengan fungsi .query()

bios.query(('height_cm > 215' and 'born_country == "USA"'))
```

```python
bios[bios['born_country'].isin(['USA','GBR']) & (bios['name'].str.startswith("Keith",na=False))]
```

# Praktek

```python
df_bios = pd.read_csv('data/bios.csv')
```

```python
df_bios.head()
```

```python
# challenge level 1
# Cari semua atlet yang memiliki tinggi badan (height_cm) lebih dari 210 cm. Tampilkan hanya kolom name dan height_cm.

df_bios.loc[df_bios['height_cm'] > 210, ['name', 'height_cm']]
```

```python
# challenge level 2
# Cari atlet asal negara Perancis (born_country == 'FRA') yang memiliki berat badan (weight_kg) di bawah 60 kg.
df_bios.query(('born_country == "FRA"') and ('weight_kg < 60'))
```

```python
# chalengge level 3
# Cari semua atlet yang namanya mengandung kata "William". Pastikan pencariannya tidak peka terhadap huruf besar/kecil (case-insensitive) dan tangani data kosong agar tidak error.

df_bios[df_bios['name'].str.contains("Williams", case=False, na=False)]
```

```python
# challenge level 4
# Filter data atlet yang lahir di negara berikut: 'USA', 'CHN' (China), dan 'JPN' (Jepang).
df_bios[df_bios['born_country'].isin(["USA","CHN","JPN"])]
```

```python
# challenge level 5
# Tantangan: Cari atlet yang nama depannya diawali dengan huruf "Z". Gunakan metode regex untuk memastikan hanya mengecek bagian awal string.
df_bios[df_bios['name'].str.contains(r'^Z', regex=True, na=False)]  # POLA PENCARIAN REGEX
```

```python
# challenge level 6
# Gunakan fungsi .query() untuk mencari atlet yang lahir di wilayah (born_region) 'California' DAN kolom tanggal kematiannya (died_date) kosong (artinya atlet tersebut masih hidup atau tidak tercatat wafat).
died = df_bios['died_date'].isnull()
df_bios.query(('born_region == "California"') and ('@died'))
```
