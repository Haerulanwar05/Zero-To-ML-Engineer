# Aggregating Data

proses menggabungkan, mengelompokkan, atau meringkas banyak baris data menjadi satu nilai ringkasan yang bermakna. Tujuan utamanya adalah untuk melihat pola besar dari dataset, bukan hanya melihat baris data satu per satu.

```python
# import library 

import pandas as pd
import numpy as np
```

```python
# Load dataframe

# Dataframe Bios
bios = pd.read_csv('data/bios.csv')

# Dataframe coffe
coffee = pd.read_csv('data/coffee_update.csv')
```

```python
# melihat jumlah orang yang lahir di tempat yang sama berdasarkan kolom 'born_city'

bios['born_city'].value_counts()
```

```python
# melihat info jumlah orang yang berada di amerika berdasarkan region yang sama
bios[bios['born_country'] == "USA"]['born_region'].value_counts()
```

```python
# melihat info jumlah orang yang berada di amerika berdasarkan region yang sama dari urutan belakang atau terkecil 

bios[bios['born_country'] == "USA"]['born_region'].value_counts().tail(25)
```
