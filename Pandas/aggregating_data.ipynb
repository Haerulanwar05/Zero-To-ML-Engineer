### Penggunaan func groupby()

Fungsi yang sangat kuat untuk mengelompokkan data berdasarkan satu atau lebih kolom, lalu menerapkan fungsi statistik pada kelompok tersebut.

```python
# menerapkan func groupby pada df coffee
coffee
```

```python
# ingin mengetahui jumlah penjualan dari tipe kopi yang dijual 

coffee.groupby(['Coffee Type'])['Units Sold'].sum()
```

```python
# ingin mengetahui rata - rata penjualan dari tipe kopi yang dijual 
coffee.groupby(['Coffee Type'])['Units Sold'].mean()
```

### menggunaka func .agg()

Memungkinkan Anda menjalankan berbagai fungsi perhitungan berbeda pada kolom yang berbeda sekaligus menggunakan kamus (dictionary). Contoh: menghitung total (sum) pada kolom penjualan dan rata-rata (mean) pada kolom harga secara bersamaan.

```python
coffee.groupby(['Coffee Type', 'Day']).agg({'Units Sold': 'sum', 'price': 'mean'})
```

### Membuat sebuah Pivot dengan func pivot atau pivot_table

```python
pivot = coffee.pivot_table(columns='Coffee Type', index='Day', values='revenue')
```

```python
pivot
```

```python
# lets say kita ingin menjumlahkan sebuah pendapatan nya 
pivot.sum()
```

```python
# melihat total pendapatan kedua jenis kopi di hari yang sama dengan parameter axis
pivot.sum(axis=1)
```

```python
# kita mengubah tipe kolom born_date menjadi sebuah tipe datetime
bios['born_date'] = pd.to_datetime(bios['born_date'], errors='coerce')
```

```python
# mengetahui jumlah orang dengan tahun lahir yang sama

# memakai func reset_index() agar menjadi table
bios.groupby(bios['born_date'].dt.year)['name'].count().reset_index()
```

```python
# mengetahui jumlah orang dengan tahun lahir dan bulan yang sama
bios['born_year'] = bios['born_date'].dt.year
bios['born_month'] = bios['born_date'].dt.month

bios.groupby([bios['born_year'], bios['born_month']])['name'].count().reset_index().sort_values('name', ascending=False)
```

Analisis Penjualan per Manufaktur (Level: Advanced)
Ini bagian paling krusial. Saya butuh tabel performa untuk setiap brand.

Tugas: Gunakan .groupby() pada kolom Manufacturer lalu gunakan fungsi .agg() untuk menghitung:

- Total penjualan (Sales_in_thousands) menggunakan sum.

- Harga tertinggi (Price_in_thousands) menggunakan max.

- Rata-rata efisiensi bahan bakar (Fuel_efficiency) menggunakan mean.

Goal: Urutkan hasilnya berdasarkan Total Penjualan terbanyak (descending).

```python
laporan_analisis = df_car.groupby('Manufacturer').agg({
    'Sales_in_thousands': 'sum',
    'Price_in_thousands': 'max',
    'Fuel_efficiency': 'mean'
    }).reset_index().sort_values(by='Sales_in_thousands', ascending=False)
```

```python
laporan_analisis
```

Saya ingin melihat brand mana yang paling seimbang antara tenaga dan keiritan.

Tugas: Kelompokkan berdasarkan Manufacturer, lalu hitung rata-rata Horsepower dan Fuel_efficiency.

Filter: Hanya tampilkan brand yang memiliki rata-rata Horsepower di atas 200.

```python
comparison_brand = df_car.groupby(['Manufacturer']).agg({
    'Horsepower' : 'mean',
    'Fuel_efficiency' : 'mean'
    }).reset_index().query('Horsepower > 200').sort_values(by='Horsepower', ascending=False)
```

```python
comparison_brand
```

Tugas 1: Agregasi Bertingkat (Agg)
Gunakan method chaining untuk:

1. Mengelompokkan data berdasarkan Manufacturer.

2. Hitung total Sales_in_thousands dan rata-rata Price_in_thousands secara bersamaan.

3. Urutkan hasilnya berdasarkan total penjualan dari yang tertinggi ke terendah.

Gunakan .reset_index() agar hasilnya menjadi DataFrame yang rapi.

```python
df_car.groupby(['Manufacturer']).agg({
    'Sales_in_thousands' : 'sum',
    'Price_in_thousands' : 'mean'
}).reset_index().sort_values(by='Sales_in_thousands', ascending=False)
```

Tugas 2: Pivot Table Multi-Dimensi
Buatlah sebuah Pivot Table yang:

1. Menjadikan Year sebagai index (baris).

2. Menjadikan Vehicle_type sebagai columns (horizontal).

3. Menghitung rata-rata dari Sales_in_thousands sebagai values-nya.

Tambahkan parameter fill_value=0 untuk menangani data yang kosong.

```python
df_car['Latest_Launch'] = pd.to_datetime(df_car['Latest_Launch'])
```

```python
df_car.info()
```

```python
df_car['year'] = df_car['Latest_Launch'].dt.year
```

```python
pivot = df_car.pivot_table(columns='Vehicle_type', index='year', values='Sales_in_thousands',aggfunc='mean', fill_value=0)
pivot
```
