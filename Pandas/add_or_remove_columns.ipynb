# Adding / Removing Columns

```python
import pandas as pd
import numpy as np
```

```python
# load dataset coffee
coffe = pd.read_csv('data/coffee.csv')
```

```python
# menampilkan 5 baris pertama dataset
coffe.head()
```

```python
# adding columns dengan nama price

coffe['price'] = 3.00
```

```python
# menghapus sebuah columns seperti contoh diatas menggantikan price dengan new_price
coffe.drop(columns=['price'])
```

```python
# jika di run ulang di coffe tidak akan terhapus karena fungsi ini menghilangkan apa yang diperintahkan tanpa menyimpan dataset baru yang sudah di edit
coffe
```

```python
# untuk bener bener mengubah nya hingga ke database maka yang perlu dilakukan yaitu menciptakan sebuah kondis dengan Inplace = True
coffe.drop(columns=['price'], inplace=True)
```

```python
# sebuah dataset baru yang disimpan setelah kita memodifikasi nya dengan menggunakan sebuah inplace = True
coffe
```

```python
'''
ada 2 cara sebenernya untuk mengubah sebuah dataset yang sudah di modifikasi dengan cara :
coffe = coffe.drop(columns=['price'])

atau 

coffe.drop(columns=['price'], inplace=True)
'''
```

```python
# kita bisa memodifikasi sebuah dataset dengan keinginan kolom apa saja yang ingin ditampilkan dalam dataset kita
coffe = coffe[['Day','Coffee Type','Units Sold','new_price']]
```

```python
coffe
```

```python
# kita ingin menambahkan sebuah kolon pendapatan
coffe['Revenue'] = coffe['Units Sold'] * coffe['new_price']

# kita bisa melakukan sebuah aritmatika untuk mengasiggning value dalam variabel kolom baru kita
```

```python
# semisal kita ingin mengganti nama kolom new_price dengan price
coffe.rename(columns = {'new_price':'price'})
```

```python
# lagii, jika kita ingin mengubah ini masuk ke dalam dataset kita harus menggunakan salah satu cara dari 2 yang tadi diatas
coffe = coffe.rename(columns = {'new_price':'price'})
```

## Praktek Adding/Removing Columns

```python
# ini merupakan sebuah praktik pribadi tanpa melihat apapun berdasarkan apa yang telah di pelajari di segmentasi ini

# load dataset coffe namun karena sudah ada dan kita tidak mau mengubah nya maka kita akan meload nya dengan variabel baru dengan menggunakan Fungsi .copy()

new_coffee = coffe.copy()
```

```python
new_coffee.head()
```

```python
# Membuat sebuah kolom baru bernama total_spent.
new_coffee['total_spent'] = new_coffee['Units Sold'] * new_coffee['price']
```

```python
new_coffee
```

```python
# menghapus kolom revenue dari new_coffe
new_coffee = new_coffee.drop(columns=['Revenue'])
```

```python
new_coffee
```

```python
# buat kolom promo_price dimana jika coffee type nya itu espresso maka diskon hingga 2.99 jika bukan maka 5.99

new_coffee['promo_price'] = np.where(new_coffee['Coffee Type'] == 'Espresso',2.99,5.99)
```

```python
new_coffee
```

```python
# mengubah nama kolom Units Sold menjadi jumlah_terjual dan kolom price menjadi harga_satuan 
new_coffee.rename(columns={'Units Sold': 'jumlah_terjual'}, inplace=True)
new_coffee.rename(columns={'price':'harga_satuan'}, inplace=True)
```

```python
new_coffee
```

```python
# menghapus kolom total_spent
new_coffee.drop(columns=['harga_satuan'],inplace=True)
```

```python
new_coffee
```

```python
# membuat sebuah kolom status_penjualan 

new_coffee['status_penjualan'] = np.where(new_coffee['jumlah_terjual'] > 25, "Target Achieved", "Under Performance")
```

```python
new_coffee
```

```python
# membuat sebuah kolom tax_amount dengan isi 10% dari pajak
new_coffee['tax_amount'] = new_coffee['promo_price'] * 10/100
```

```python
new_coffee
```

```python
# mengubah kolom Coffee type menjadi jenis_kopi

new_coffee.rename(columns={'Coffee Type':'jenis_kopi'}, inplace=True)
```

```python
new_coffee
```

```python
# membuat sebuah kolom baru bernama final_price dengan isi promo_price + tax_amount
new_coffee['final_price'] = new_coffee['promo_price'] + new_coffee['tax_amount']
```

```python
new_coffee
```
