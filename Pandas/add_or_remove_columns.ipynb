## Bios Dataset for add/remove column example

```python
df_bios = pd.read_csv('data/bios.csv')
```

```python
# menampilkan 5 baris pertama data
df_bios.head()
```

```python
# membuat sebuah kolom dengan label 'first name' yang berfungsi mengisi value nama awal nya saja

# membuat sebuah variabel dengan isi copy an variabel yang sudah di load dataset agar tidak berubah dataset di dalam variabel tersebut

bios_new = df_bios.copy()
```

```python
# membuat sebuah kolom dengan varibel copy an dari variabel dataset asli
bios_new ['first_name'] = bios_new['name'].str.split(' ').str[0]
```

```python
bios_new.head()     # menampilkan 5 data pertama
```

```python
# mencari nama dengan fungsi query dimana first_name nya yaitu Keith

bios_new.query('first_name == "Keith"')
```

```python
bios_new.info()
```

```python
# untuk berurusan dengan sebuah data seperti kolom born_date dimana seharusnya dia sebuah angka atau date kita bisa mengubah nya seperti ini
bios_new['born_datetime'] = pd.to_datetime(bios_new['born_date'])
```

```python
bios_new
```

```python
# tidak terlihat berbeda namun jika kita melihat dari info nya 
bios_new.info()


# dia akan beruba sebuah datetime dtype nya
```

```python
# membuat sebuah kolom dengan catatan hanya mengambil tahun dari kelahiran nya saja dengan fungsi pandas yaitu (.dt) ini sama seperti (.str) tetapi dikhususkan untuk tipe data datetime

bios_new['born_year'] = bios_new['born_datetime'].dt.year

bios_new[['name','born_year']]


# jadi fungsi (.dt) ini sangat berguna untuk mengetahui apapun tentang datetime seperti leapyear atau hari pada bulan apa dan apapun
```

Parameter Penting pd.to_datetime():

errors='coerce': Menangani data tidak valid (seperti teks "Unknown") dengan mengubahnya menjadi NaT (Not a Time). Tanpa ini, program akan error jika bertemu data rusak.

format=: Menentukan pola tanggal (misal: %d-%m-%Y untuk 31-12-2024). Penggunaan format spesifik dapat meningkatkan kecepatan proses konversi secara signifikan.


Mengelompokkan data numerik height_cm ke dalam 3 kategori label:

Short: Tinggi di bawah 165 cm.

Average: Tinggi antara 165 cm hingga kurang dari 185 cm.

Tall: Tinggi 185 cm ke atas.

```python
# mengkategorikan seseorang berdasarkan tinggi nya [short,average,tall]
df_bios['height_category'] = df_bios['height_cm'].apply(lambda x: 'Short' if x < 165 else ('Average' if x < 185 else 'Tall'))
```

```python
df_bios.head()
```

```python
# mengkategorikan class atlet seperti [lightweight,middleweight,heavyweight]

# tetapi ini dengan membuat sebuah fungsi 

def categorize_athlete (row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'lightweight'
    elif row['height_cm'] < 185 or row['weight_kg'] <= 80:
        return 'middleweight'

    else:
        return 'heavyweight'

df_bios['categorize'] = df_bios.apply(categorize_athlete, axis=1)
```

## Praktek Adding/Removing Columns

```python
# ini merupakan sebuah praktik pribadi tanpa melihat apapun berdasarkan apa yang telah di pelajari di segmentasi ini

# load dataset coffe namun karena sudah ada dan kita tidak mau mengubah nya maka kita akan meload nya dengan variabel baru dengan menggunakan Fungsi .copy()

new_coffee = coffe.copy()
```

```python
new_coffee.head()
```

```python
# Membuat sebuah kolom baru bernama total_spent.
new_coffee['total_spent'] = new_coffee['Units Sold'] * new_coffee['price']
```

```python
new_coffee
```

```python
# menghapus kolom revenue dari new_coffe
new_coffee = new_coffee.drop(columns=['Revenue'])
```

```python
new_coffee
```

```python
# buat kolom promo_price dimana jika coffee type nya itu espresso maka diskon hingga 2.99 jika bukan maka 5.99

new_coffee['promo_price'] = np.where(new_coffee['Coffee Type'] == 'Espresso',2.99,5.99)
```

```python
new_coffee
```

```python
# mengubah nama kolom Units Sold menjadi jumlah_terjual dan kolom price menjadi harga_satuan 
new_coffee.rename(columns={'Units Sold': 'jumlah_terjual'}, inplace=True)
new_coffee.rename(columns={'price':'harga_satuan'}, inplace=True)
```

```python
new_coffee
```

```python
# menghapus kolom total_spent
new_coffee.drop(columns=['harga_satuan'],inplace=True)
```

```python
new_coffee
```

```python
# membuat sebuah kolom status_penjualan 

new_coffee['status_penjualan'] = np.where(new_coffee['jumlah_terjual'] > 25, "Target Achieved", "Under Performance")
```

```python
new_coffee
```

```python
# membuat sebuah kolom tax_amount dengan isi 10% dari pajak
new_coffee['tax_amount'] = new_coffee['promo_price'] * 10/100
```

```python
new_coffee
```

```python
# mengubah kolom Coffee type menjadi jenis_kopi

new_coffee.rename(columns={'Coffee Type':'jenis_kopi'}, inplace=True)
```

```python
new_coffee
```

```python
# membuat sebuah kolom baru bernama final_price dengan isi promo_price + tax_amount
new_coffee['final_price'] = new_coffee['promo_price'] + new_coffee['tax_amount']
```
